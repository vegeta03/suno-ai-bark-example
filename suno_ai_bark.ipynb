```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio

# download and load all models
preload_models()

# generate audio from text
text_prompt = """
     [WOMAN] Hello, my name is Suno. And, uh — Hi Shyam. [laughs] I like to get to know you more. — Maybe we can meet in person.
"""

audio_array = generate_audio(text_prompt, history_prompt="v2/en_speaker_9")
# save audio to disk
filename = "Shyam bark_generation2.wav"
write_wav(filename, SAMPLE_RATE, audio_array)
# play text in notebook
Audio(audio_array, rate=SAMPLE_RATE, filename=filename)

```
